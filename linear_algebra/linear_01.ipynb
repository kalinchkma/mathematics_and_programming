```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Linear Algebra

```python

# vector x and y
x = np.array([[1],[5]])

y = np.array([[1],[17]])

z = x + y

# plot
fig, ax = plt.subplots()

ax.plot(x)
ax.plot(y)
ax.plot(z)

plt.grid(False)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the points for the geometric plot
x = [1, 2, 3, 4, 1]  # X-coordinates of the points
y = [1, 3, 2, 4, 1]  # Y-coordinates of the points

# Create a new figure and axis object
fig, ax = plt.subplots()

# Plot the geometric plot
ax.plot(x, y, marker='o', linestyle='-')

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Geometric Plot')

# Display the plot
plt.grid(True)
plt.show()
```
